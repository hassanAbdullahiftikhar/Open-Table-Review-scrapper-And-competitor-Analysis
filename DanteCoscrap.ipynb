```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.common.exceptions import StaleElementReferenceException
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url="https://www.opentable.com/r/donato-and-co-berkeley?corrid=d96626a2-c2d9-420e-8796-b171698d9628&avt=eyJ2IjoyLCJtIjoxLCJwIjowLCJzIjowLCJuIjowfQ&p=2&sd=2024-11-27T23%3A00%3A00"
driver.get(url)
import time
from selenium.common.exceptions import ElementNotInteractableException, NoSuchElementException,ElementClickInterceptedException
data=[]
dates=[]
rating=[]
while(True):
    time.sleep(2)
    driver.execute_script("window.scrollTo(0,500)")
    rew=driver.find_elements(By.CLASS_NAME,"_6rFG6U7PA6M-")
    date=driver.find_elements(By.CLASS_NAME,"iLkEeQbexGs-")
    ratings=driver.find_elements(By.CLASS_NAME,"-k5xpTfSXac-")
    if not rew:
            print("No reviews retrieved. Refreshing the page...")
            driver.refresh()
            time.sleep(5)  # Wait for the page to reload
            continue 
    for i in rew:
        data.append(i.text)
    for i in date:
         dates.append(i.text)
    for i in ratings:
         rating.append(i.text)
    button=driver.find_element(By.XPATH,"/html/body/div[1]/div/div/main/div/div[2]/div[1]/section[7]/section/footer/div/div[2]")
    try:button.click()
    except NoSuchElementException:
        time.sleep(3)
        button=driver.find_element(By.XPATH,"/html/body/div[1]/div/div/main/div/div[2]/div[1]/section[7]/section/footer/div/div[2]")
        button.click()
    except StaleElementReferenceException:
        time.sleep(3)
        button=driver.find_element(By.XPATH,"/html/body/div[1]/div/div/main/div/div[2]/div[1]/section[7]/section/footer/div/div[2]")
        button.click()
    except ElementClickInterceptedException:
        driver.refresh()
        time.sleep(3)
    except ElementNotInteractableException:
        break
    
    
```

```python
import pandas as pd
data=pd.DataFrame(data)
dates=pd.DataFrame(dates)
data.columns=['Reviews']
dates.columns=['Date']
```

```python
over=rating[::4]
food=rating[1::4]
serv=rating[2::4]
amb=rating[3::4]
food=[int(i.split()[1])for i in food]
over=[int(i.split()[1])for i in over]
serv=[int(i.split()[1])for i in serv]
amb=[int(i.split()[1])for i in amb]
over=pd.DataFrame(over,columns=["Overall Rating"])
food=pd.DataFrame(food,columns=["Food"])
serv=pd.DataFrame(serv,columns=["Service"])
amb=pd.DataFrame(amb,columns=["Ambiance"])
```

```python
review=pd.concat([data,dates,over,food,serv,amb],axis=1)
```
