```python
import anthropic
import pandas as pd
def extract_review(prompt,index):
    try:
        client = anthropic.Anthropic()
        message = client.messages.create(
            model="claude-3-5-sonnet-20241022",
            max_tokens=1000,
            temperature=0,
            system="You are required to extract the information about the quality of food and service provided.Keep in mind to not include any personal information of the reviewer and ignore any other feedback plus dont write any additional notes just provide the extracted review and nothing more,indicate food review like 1.(food review here)\n2.(service review here)",
            messages=[
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "text",
                            "text": prompt
                        }
                    ]
                }
            ]
        )
        d=[t.text for t in message.content][0]
        print(index,d)
        return d
    except Exception as e:
        print(e)
```

```python

df=pd.read_excel("Reviewdata.xlsx")
r=df.loc[:,"Reviews"]
extracted_data=[extract_review(s,i) for i,s in enumerate(r)]
```

```python
import numpy as np
categ=np.array(extracted_data,dtype=str)
split_reviews = [entry.split("2.") for entry in categ]
food_reviews = [review[0].strip() if len(review) > 0 else "" for review in split_reviews]
service_reviews = [review[1].strip() if len(review) > 1 else "" for review in split_reviews]
food_reviews = [review[3:] if review.startswith('1.') else review for review in food_reviews]
```
